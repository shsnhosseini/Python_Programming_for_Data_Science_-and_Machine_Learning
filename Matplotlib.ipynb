### 2. Matplotlib Library



*   Matplotlib is a versatile library for **data visualization** in Python, offering a wide range of plotting options and customization capabilities. Whether you're working on simple line plots or complex interactive visualizations, Matplotlib provides the tools needed to effectively communicate your data insights.

* Matplotlib is a fundamental Python library for creating static, animated, and interactive visualizations of data. It excels in generating various plot types, including:

  *   Line plots (ideal for trends over time)
  *   Scatter plots (useful for exploring relationships between variables)

  *   Bar charts (effective for comparing categories)
  *   Histograms (valuable for understanding data distribution)
  *   Pie charts (suitable for proportions of a whole)
  *   Heatmaps (great for visualizing data with two categorical variables)
  *   3D plots (helpful for representing complex relationships)

*   Matplotlib integrates seamlessly with other popular Python libraries for IoT development, such as:

    * NumPy: Provides efficient numerical computation capabilities.
    * Pandas: Offers data structures (Series, DataFrames) for handling large datasets.
    * SciPy: Extends scientific computing functionalities.


*   In the context of **Internet of Things (IoT)** applications, Matplotlib becomes particularly useful for visualizing data collected from sensors and devices. This visualization can range from simple line graphs showing temperature changes over time to complex multi-dimensional plots representing various sensor readings. The ability to quickly generate these visualizations helps in understanding patterns, making predictions, and debugging issues within IoT systems.

```python
# Example 1: Visualizing Sensor Data
""" Let's consider a simple example where we have temperature data collected from a sensor over time,
 and we want to visualize this data using Matplotlib."""

 # Sample data
time_stamps = ['10:00', '11:00', '12:00', '13:00', '14:00', '15:00']
temperature = [22, 23, 23.5, 25, 27, 27.3]

plt.figure(figsize=(4, 4))
plt.plot(time_stamps, temperature, 'o-.c', label='Temperature')  # plot() is a function within the pyplot module.
plt.title('Temperature Over Time', fontsize=10, color='b', loc='center')  # title() is a function within the pyplot module.
plt.xlabel('Time')  # xlabel() is a function within the pyplot module.
plt.ylabel('Temperature (°C)')   # ylabel() is a function within the pyplot module.
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability if many
plt.grid(True)  # Add grid lines for easier interpretation
plt.legend(loc='upper center', shadow=True)  # legend() function is used to add a legend to your plot
plt.show()  # show() is a function within the pyplot module. It's used to display all currently active figure objects in interactive windows.
```

**subplot**

  *   In the realm of Matplotlib for IoT applications, plt.subplot is a valuable tool for creating multiple plots within
a single figure window `simultaneously`.

  *   This is particularly useful when you want to visualize different aspects of your sensor data or compare data from multiple devices on the same screen.

  * For example, when dealing with connected cars, data visualization becomes crucial for understanding various parameters such as speed, acceleration, fuel efficiency, and sensor readings over time. Using plt.subplot, engineers and analysts caneffectively display this information in an organized manner, facilitating better analysis and decision-making processes.


```python
# Simulated data for connected car
time = np.linspace(0, 10, 100)
sensor_1 = np.sin(time) * 20 + 50
sensor_2 = np.cos(time) * 5 + 15
sensor_3 = np.random.normal(0, 1, 100)

plt.figure(figsize=(15, 5))

# Create a figure with 3 subplots (1 row, 3 columns)
# Subplot 1: Sensor_1
plt.subplot(1, 3, 1)
plt.plot(time, sensor_1, 'r-')
plt.title('Sensor 1')
plt.xlabel('Time(s)')
plt.ylabel('Sensor Value')

# Subplot 2: sensor_2
plt.subplot(1, 3, 2)
plt.plot(time, sensor_2, 'g-')
plt.title('Sensor 2')
plt.xlabel('Time(s)')
plt.ylabel('Sensor Value')

# Subplot 3: Sensor_3
plt.subplot(1, 3, 3)
plt.plot(time, sensor_3, 'b-')
plt.title('Sensor 3')
plt.xlabel('Time(s)')
plt.ylabel('Sensor Value')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```

**Histogram**:

  * A histogram is a type of chart that graphically represents the **distribution** of numerical data by grouping the data into **bins** or intervals and displaying the **frequency** of data points within each bin as vertical bars. Here are some key points about histograms:

    * **Structure**:

        X-axis: Represents the range of data values divided into bins.

        Y-axis: Represents the frequency or count of data points within each bin.

   *  **Purpose**:

        Histograms are used to depict the distribution of a dataset, showing how often values fall into specific ranges.

        They help to visualize the shape, spread, and central tendency of the data.

   * **Differences from Bar Charts**:

        While both histograms and bar charts use vertical bars, histograms are used for **continuous data** divided into intervals, whereas bar charts are used for **categorical data**.

   * **Customization**:

        The number of bins and their intervals can be adjusted to change the granularity of the histogram.

        Histograms can be customized to show **frequencies**, **percentages**, or **densities**.


```python
""" Histograms are particularly useful in connected car data analysis for visualizing distributions of various metrics like
 speed, fuel efficiency, or sensor readings, helping to identify patterns or anomalies in the data."""

# Simulating speed data for a connected car
speeds = np.random.normal(80, 10, 1000)  # Mean 60 km/h, std dev 10 km/h

plt.figure(figsize=(7, 4))
plt.hist(speeds, bins=200, edgecolor='black', color='c')
plt.title('Distribution of Vehicle Speeds')
plt.xlabel('Speed (km/h)')
plt.ylabel('Frequency') # frequency: count of data points within each bin
plt.grid(True, alpha=0.3)
plt.show()
```

**Question:** Why does the distribution change every time when the code is executed?

```python
# solution
# Set a fixed seed for reproducibility
# By setting this seed before generating your random numbers, you ensure that the same sequence of "random" numbers is
# generated each time you run the script.
np.random.seed(1403)  # You can use any integer value as the seed
```

**3D Plotting**

```python
# basic setup
import matplotlib.pyplot as plt
import numpy as np
```

1. 3D Scatter Plot

```python
# creating an object with name fig
fig = plt.figure()

# Create a 3D subplot object
ax = fig.add_subplot(projection='3d')

# Define data points
x = np.array([1, 3, 3, 4, 5])
y = np.array([3, 4, 6, 8, 10])
z = np.array([2, 5, 20, 7, 9])

# distance from the origin
distance = np.sqrt(x**2 + y**2 + z**2)

# Calculate the size based on distance (normalized)
size = (distance / distance.max() ) * 100

# Create a scatter plot
# changing the size of the dots with the s argument
ax.scatter(x, y, z, s=size)


# Set labels and title
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label', labelpad=2)
ax.set_title('3D Scatter Plot Example')

plt.show()
```

```python
# Create a new figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

#  Generate random data for the scatter plot
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

# coloring each point individually based on some criteria
# each point's color in the scatter plot is determined by its corresponding value in c
# if c is a single scalar value, all points will be colored with the same color according to the colormap specified.
c = x + y + z    # You can change this to any function of x, y, z

# You can modify the c variable to reflect different criteria, such as distance from the origin.
sc = ax.scatter(x, y, z, c=c, cmap='coolwarm', s=30)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# Add a colorbar
cbar = plt.colorbar(sc, shrink=0.5, pad=0.15)
cbar.set_label('Colorbar Label')

# Add a colorbar
plt.show()
```

2.  3D Line Plot

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
z = [1, 3, 13, 7, 9]

ax.plot3D(x, y, z, 'y--*', linewidth = '1.5')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label',labelpad=-1)
ax.set_title('3D Line Plot')

plt.show()
```

3. 3D Surface Plot

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
x, y = np.meshgrid(x, y)
z = x**2 + y**2

ax.plot_surface(x, y, z)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label', labelpad=2)
ax.set_title('3D Surface Plot')

plt.show()
```

4. 3D Wireframe Plot

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10, color='c')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z', labelpad=2)
ax.set_title('3D Wireframe Plot')

plt.show()
```

5. 3D Bar Plot

```python
from matplotlib import cm
from matplotlib.colors import Normalize

# Create a new figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Generate data for the bar plot
x = np.arange(5)
y = np.arange(5)
X, Y = np.meshgrid(x, y)
Z = np.random.rand(5, 5)  # Random heights for the bars

# Normalize the Z values for colormap
norm = Normalize(vmin=Z.min(), vmax=Z.max())
colors = cm.viridis(norm(Z.ravel()))  # Apply colormap

# Create the 3D bar plot
bars = ax.bar3d(X.ravel(), Y.ravel(), np.zeros_like(Z.ravel()),
                 0.8, 0.8, Z.ravel(), color=colors)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Bar Plot with Colormap')

# Create a colorbar based on the normalization and colormap
mappable = cm.ScalarMappable(cmap=cm.viridis, norm=norm)
fig.colorbar(mappable, ax=ax, shrink=0.5, pad=0.15)

# Show the plot
plt.show()
```

2D Contour Plot

```python
plt.figure(figsize=(4, 3))
# Create a 2D grid of x and y values
x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x, y)

# Define a function to plot
Z = X**2 + Y**2

# Create the contour plot
contour = plt.contour(X, Y, Z, levels=20, cmap='coolwarm')

# Add a color bar
plt.colorbar(contour, label='Z')

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2D Contour Plot')

# Save the plot to a file
# DPI (Dots Per Inch): The dpi parameter controls the resolution of the saved figure.
# The bbox_inches='tight' option adjusts the bounding box of the saved figure to include all elements of the plot, ensuring that labels and titles are not cut off.
plt.savefig('contour_plot.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```
