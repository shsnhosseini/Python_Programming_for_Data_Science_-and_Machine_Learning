Pearson Correlation Coefficient



*   Pearson’s correlation coefficient(r) is a widely used metric for assessing linear relationships between two random variables
*   In the Context of DataFrames:
    * When we use .corr() on a DataFrame, each column is treated as a realization of a random variable.
    * The method calculates the sample correlation coefficient, which is an **estimate** of the population
    correlation coefficient.

*   Sample vs. Population:

    * In practice, we usually work with sample data, so we calculate the sample Pearson correlation coefficient(r).
    * This is an estimate of the population correlation coefficient(ρ).

*   Assumptions:

    * The Pearson correlation assumes that the relationship between the variables is linear.
    * It also assumes that the variables are normally distributed (for inferential purposes).


*   Interpretation:

    * The coefficient ranges from -1 to 1.
    * A value of 1 implies that a linear equation perfectly describes the relationship between X and Y, with all data points lying on a line for which Y increases as X increases.
    * A value of −1 implies that all data points lie on a line for which Y decreases as X increases.
    * A value of 0 implies that there is no linear correlation between the variables(**uncorrelated**)

```python
# Heatmap of Correlation between Temperature and Humidity:
# Calculating the correlation matrix:
# By default, .corr() uses the Pearson correlation coefficient, which measures the linear correlation between two random variables
# corr() method to compute the correlation matrix.
# By default, it computes the Pearson correlation coefficient, but other methods like Spearman
# and Kendall can also be used depending on the nature of the data.
# the .corr() method in pandas DataFrames is used to compute the pairwise correlation of columns

plt.figure(figsize=(5, 3))
correlation_matrix = df[['temperature', 'humidity']].corr(method='pearson')
print(correlation_matrix)

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.1f')
plt.title('Correlation Heatmap: Temperature vs. Humidity', fontsize=10)
plt.show()
```

GroupBy

*    The groupby method allows users to split data into groups, apply a function to each group independently, and then combine the results.

*    This process is often referred to as the "split-apply-combine" strategy.

```python
# Sample DataFrame representing IoT telecom data
data = {
    'DeviceID': ['device1', 'device2', 'device1', 'device2', 'device1', 'device2', 'device1', 'device1'],
    'DataType': ['temperature', 'humidity', 'pressure', 'temperature', 'humidity', 'pressure', 'temperature', 'pressure'],
    'DataUsageMB': [10, 15, 20, 25, 30, 35, 40, 45]  # Data usage in megabytes
}

df_IoT = pd.DataFrame(data)
print("Original IoT Telecom Data:")
print(df_IoT)

# Group by 'DeviceID' and aggregate the sum of 'DataUsageMB'
result = df_IoT.groupby('DeviceID')['DataUsageMB'].sum()

print("\n\nTotal Data Usage by each Device:")
print(result)
```

Pivot Table

*   Pandas provides methods for manipulating a Series and DataFrame to alter the **representation** of the data for further data processing or data summarization.



```python
# Sample telecom IoT data
data = {
    'DeviceID': ['Device1', 'Device1', 'Device2', 'Device3', 'Device1', 'Device2', 'Device3'],
    'Region': ['North', 'North', 'South', 'East', 'West', 'North', 'South'],
    'DataUsageMB': [500, 1000, 2000, 1500, 3000, 2500, 3500],
    'Month': ['Jan', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
}
df_1 = pd.DataFrame(data)

print(df_1)

# Pivot table to get average data usage per device per month across different regions
# This gives insights into how much data each device is using on average in each region over time.
pivot_df = df_1.pivot_table(index=['DeviceID', 'Month'], columns='Region', values='DataUsageMB', aggfunc='mean')

print(f'\n\n {pivot_df}')
```

```python
pivot_df_ev = df.pivot_table(index=['Make', 'Model'], columns='Electric Vehicle Type', values='Electric Range', aggfunc='mean')
pivot_df_ev.shape
pivot_df_ev.to_excel('pivot_df_ev.xlsx')
```

```python
# defining custom aggregation functions
pivot_df_ev_custom = df.pivot_table(index=['Make', 'Model'], columns='Electric Vehicle Type', values='Electric Range', aggfunc=lambda x: x.sum() + x.mean())
pivot_df_ev_custom.to_excel('pivot_df_ev_custom.xlsx')
```

```python
# calculating Multiple Values
pivot_df_ev_multiple_values = df.pivot_table(index=['Make', 'Model'], columns='Electric Vehicle Type',
                                             values=['Electric Range', 'Model Year'],
                                             aggfunc={'Electric Range': 'mean', 'Model Year': 'min'}
                                             )
pivot_df_ev_multiple_values.to_excel('pivot_df_ev_multiple.xlsx')
```

**Key Differences between GroupBy and Pivot Table**

*   **Shape of Output**: groupby results in a long, stacked format, while pivot_table results in a wide, matrix-like format.
*   **Complexity**: pivot_table can handle more complex data transformations by allowing multiple levels of aggregation and cross-tabulation, whereas groupby is more straightforward and linear.

#### Memory Management and Speed Optimization

*  **Memory Management**

  * **1. Use Memory-Efficient Data Structures**: Instead of using large data structures, opt for more memory-efficient ones. For instance, consider using NumPy arrays instead of Pandas DataFrames if you don't need the additional functionality provided by Pandas.

  * **2. Optimize Data Loading**: Load only necessary data into memory. If you're dealing with large datasets, use Pandas' read_csv() with parameters like **usecols** to load only specific columns.

  *  **3. Garbage Collection**: Python's garbage collector can be manually invoked to free up memory. Use **gc.collect()** to clean up unused objects.

  *  **4. Data Types Optimization**: Use the appropriate data types for your DataFrame columns. For example, use astype('category') for columns with a limited number of unique values.



* **Speed Optimization**

  *  **1. Vectorization**: Use Pandas' built-in vectorized operations instead of looping through DataFrame rows. This can significantly speed up data manipulation tasks.

  *  **2. Use of Efficient Libraries**: Libraries like NumPy and Pandas are optimized for performance. Ensure you're using the latest versions to benefit from performance improvements.
